```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic Plot of a and y values
x = [1,2,3,4,5]
y = [123,7,34,2,0]

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('y')
plt.grid('True')
plt.title('Basic Plot')

plt.show()
```

```python
# Sine Graph
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.title('Sine Graph')

x = np.linspace(-10000 * np.pi, 10000 * np.pi, 1000)
y = np.sin(x)
y1 = np.sin(x * x)
y2 = np.sin(x * x * x)
y3 = np.sin(x * x * x * x)
y4 = np.sin(x * x * x * x * x)
plt.plot(x, y, 'red', x, y1, 'blue', x, y3, 'black')
plt.show()
```

```python
# Cosine Graph
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.title('Cosine Graph')

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.cos(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-100, 100, 180)
y = np.linspace(-100, 100, 180)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Compute the values of sin(x^2 + y^2) and cos(xy)
Z1 = np.sin(X**2 + Y**2)
Z2 = np.cos(X * Y)

# Plot the contour lines where sin(x^2 + y^2) = cos(xy)
plt.contour(X, Y, Z1 - Z2, levels=[0], colors='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of sin(x^2 + y^2) - cos(xy) = 0')
plt.grid(True)
plt.show()
```
